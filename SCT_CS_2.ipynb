```python
!sudo apt install tesseract-ocr -y
!pip install pytesseract

from PIL import Image
import pytesseract
import io

# Caesar Cipher Decryption Function
def caesar_decrypt(text, key):
    decrypted = ''
    for char in text:
        if char.isupper():
            decrypted += chr((ord(char) - key - 65) % 26 + 65)
        elif char.islower():
            decrypted += chr((ord(char) - key - 97) % 26 + 97)
        else:
            decrypted += char
    return decrypted

# Upload Image
from google.colab import files
uploaded = files.upload()
image_name = list(uploaded.keys())[0]
image = Image.open(io.BytesIO(uploaded[image_name]))

# Extract text using pytesseract
extracted_text = pytesseract.image_to_string(image)

# Clean text: remove extra line breaks and join into one line
cleaned_text = ' '.join(extracted_text.split())

print("🧾 Extracted Text (One Line):\n", cleaned_text)

# Enter Caesar key and decrypt
key = int(input("🔑 Enter Caesar Cipher Key to Decrypt: "))
decrypted_text = caesar_decrypt(cleaned_text, key)

print("\n✅ Decrypted Text:\n", decrypted_text)

# Save decrypted result
```
